```python
import numpy as np

class OptimalBinarySearchTree:    
        
    def __init__(self, keys, probab_keys, probab_pseudokeys = None):        
        self.probab_claves = probab_keys
        if probab_pseudokeys == None:
            probab_pseudokeys=[0]*(len(probab_keys)+1)
        self.probab_pseudo = probab_pseudokeys
        self.claves = keys
        self.tabla_raiz = {}
        self.e = {}
        self.w = {}
        self.nodo_raiz = ""
        self.crear_tablas()        
        
    def crear_tablas(self):        
        n = len(self.probab_claves)
        
        for i in range(1, n + 2):
            self.e[(i, i - 1)] = self.probab_pseudo[i - 1]
            self.w[(i, i - 1)] = self.probab_pseudo[i - 1]
            
        for l in range(1, n + 1):
            for i in range(1, n - l + 2):
                j = i + l - 1
                self.e[(i, j)] = float("inf")
                self.w[(i, j)] = self.w[(i, j - 1)] + self.probab_claves[j - 1] + self.probab_pseudo[j]
                for r in range(i, j + 1):
                    t = round(self.e[(i, r - 1)] + self.e[(r + 1, j)] + self.w[(i, j)], 2)
                    if t < self.e[(i, j)]:
                        self.e[(i, j)] = t
                        self.tabla_raiz[(i, j)] = r
         
        self.nodo_raiz = self.claves[self.tabla_raiz[1,len(self.claves)]-1]
    
    def __len__(self):        
        return len(self.claves)        
        
    def __contains__(self, key):        
        return (key in self.claves)
        
    def __getitem__(self, i):        
        return self.claves[i]    

    def __str__(self, key = None):       
        if key == None:
            key = self.root()
            
        return self.string_tree(key)
          
    def string_tree(self, key = None):
        if key == None:
            return ""
        
        hijos = self.hijos(key)
        tree = "(" + self.string_tree(hijos[0]) + key + self.string_tree(hijos[1]) + ")"        
        return tree
            
    def hijos(self, key = None):        
        if key == None:            
            key = self.nodo_raiz
            
        indice_raiz = self.claves.index(self.nodo_raiz)+1        
        indice_izquierdo = (1, indice_raiz-1)        
        indice_derecho = (indice_raiz + 1, len(self.probab_claves))
        nodo_derecho = self.claves[self.tabla_raiz[indice_derecho]-1]
        nodo_izquierdo = self.claves[self.tabla_raiz[indice_izquierdo]-1]
        
        if (key != self.nodo_raiz):
            claveActual = self.nodo_raiz
            while claveActual != key:
                               
                if self.claves.index(claveActual) > self.claves.index(key):
                    i,j = indice_izquierdo                   
                else:
                    i,j = indice_derecho     
                    
                claveActual = self.claves[self.tabla_raiz[i,j]-1]                
                raiz = self.claves.index(claveActual)+1                
                
                indice_izquierdo = (i, raiz - 1)
                indice_derecho = (raiz + 1, j)                
                
                if indice_izquierdo[1]>=indice_izquierdo[0] and indice_izquierdo in self.tabla_raiz:
                    nodo_izquierdo = self.claves[self.tabla_raiz[indice_izquierdo]-1] 
                else:
                    nodo_izquierdo = None
                
                if indice_derecho[1]>=indice_derecho[0] and indice_derecho in self.tabla_raiz:
                    nodo_derecho = self.claves[self.tabla_raiz[indice_derecho]-1] 
                else:
                    nodo_derecho = None                                    
                                
        return (nodo_izquierdo, nodo_derecho)

    def depth(self, key = None):                            
        if key == None:
            return self.lenght(self.root())
        elif key not in self.claves:
            return None
        else:
            depth = 0
            actual = self.root()

            while (actual != key):
                depth = depth + 1
                hijos = self.hijos(actual)
                    
                if self.claves.index(key) > self.claves.index(actual):
                    actual = hijos[1]
                else:
                    actual = hijos[0]

            return depth
        
    def lenght(self, key):        
        if key == None:
            return -1
        else:
            hijos = self.hijos(key)            
            return 1 + max(self.lenght(hijos[0]), self.lenght(hijos[1]))

    def root(self):        
        return self.nodo_raiz        
        
    def cost(self, key=None):       
        if key == None or key == self.root():
            return self.e[1, len(self.claves)]
        else:
            #TODO
            print('TODO')
            return 1
        
    def print(self, datos, longitud):
        matriz = np.zeros((longitud+1, longitud+1))
        for fila, columna in datos:
            matriz[fila-1][columna-1] = datos[fila, columna] 

        i = 0
        for fila in matriz:
            longitud = len(fila)-1        
            temp = np.insert(fila, 0, fila[longitud])        
            temp = np.delete(temp, longitud+1)        
            fila = temp        
            matriz[i] = fila
            i+=1
        print(matriz)        
        print()        
```
