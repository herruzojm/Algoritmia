```python
def intercambia_elemento(lista, elemento1, elemento2):
    print('vamos a cambiar el', elemento1, '(',lista[elemento1],') por el', elemento2, '(',lista[elemento2],')')
    aux = lista[elemento1]
    lista[elemento1] = lista[elemento2]
    lista[elemento2] = aux
    print(lista)
    print()
    return lista

def particion_bidireccional(lista, posicion_pivote=0):
    alist = list(lista)
    primero = 0
    ultimo = len(lista) - 1
    pivote = lista[posicion_pivote]

    while primero < ultimo:
        while alist[primero] < pivote:
            primero += 1

        while alist[ultimo] > pivote:
            ultimo -= 1

        if primero <= ultimo:
            alist = intercambia_elemento(alist, primero, ultimo)
            primero += 1
            ultimo -= 1

    return alist

def particion_unidireccional(lista, posicion_pivote=0):
    alist = list(lista)
    pivote = lista[posicion_pivote]
    for i in range(1, len(alist)):
        if alist[i] < pivote:
            posicion_pivote += 1
            alist = intercambia_elemento(alist, posicion_pivote, i)    
        
    return intercambia_elemento(alist, 0, posicion_pivote)



if __name__ == '__main__':
    lista = [69,21,23,90,4,36,88,2,2,81,95]
    #lista = [42, 90, 5, 76, 37, 47, 2, 37, 46, 6, 11 ]
    print('LISTA ORIGINAL \n{}'.format(lista))    
    print()
    print("LISTA BIDIRECCIONAL")
    resultado = particion_bidireccional(lista, 0)
    print('RESULTADO-->',resultado)
    print()
    print("LISTA UNIDIRECCIONAL")
    resultado = particion_unidireccional(lista, 0)
    print('RESULTADO-->',resultado)
```
