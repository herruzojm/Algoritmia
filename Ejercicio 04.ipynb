```python
def raice_counter(counter, base = 2):    
    indice = len(counter) 
    elementos_cambiados = 0
    while True:
        indice = indice - 1
        counter[indice] = counter[indice] + 1 if counter[indice] + 1 < base else 0
        elementos_cambiados += 1              
        if counter[indice] != 0 or indice == 0:
            break
        
    return counter, elementos_cambiados
    
```

```python
import collections
class wallet:   
    def __init__(self, notes = None):             
        self._billetes = {}
        self += notes
        
    def value(self):        
        suma = 0
        for valor, cantidad in self._billetes.items():
            suma += valor*cantidad
        return suma

    def notes_count(self, note_value = None):        
        if note_value:
            if note_value in self:
                return self[note_value]
            else:
                return 0
        else:
            return sum(self._billetes.values())
    
    def __contains__(self, note):
        return note in self._billetes
    
    def __getitem__(self, note):
        if note in self:
            return self._billetes[note]
        else:
            return 0        

    def __setitem__(self, note, amount):
        if amount > 0:
            self._billetes[note] = amount
        else:
            del self[note]              
                
    def __delitem__(self, note):        
        if note in self:
            del self._billetes[note]
    
    def __eq__(self, other):
        iguales = True
        for valor, cantidad in self._billetes.items():
            if other[valor] != cantidad:
                iguales = False
                break            
        return iguales
        
    def __add__(self, notes):
        nueva_cartera = wallet(self)
        nueva_cartera+=notes
        return nueva_cartera
      
    def __radd__(self, other):        
        if other:
            return self.__add__(other)
        else:
            return self
        
    def __iadd__(self, notes):
        if notes:
            if isinstance(notes, wallet):
                for valor, cantidad in notes:
                    self[valor] = self[valor] + cantidad
            elif isinstance(notes, dict):                 
                for valor, cantidad in notes.items():                    
                    self[valor] = self[valor] + cantidad
            elif isinstance(notes, collections.Iterable):
                for valor in notes:
                    self[valor] = self[valor] + 1
            else:
                self[notes] = self[notes] + 1
                    
        return self
        
    def __iter__(self):
        return iter(self._billetes.items())            
```

```python
def coin_change(amount, notes):     
    billetera = {}
    if isinstance(notes, wallet):
        for valor, numero in notes:
            billetera[valor] = numero
    elif isinstance(notes, collections.Iterable):
        for valor in notes:
            if valor in billetera:
                billetera[valor] += 1
            else:
                billetera[valor] = 1
    else:
        billetera[notes] = 1
        
    
    cambio = {}
    tipos_billete = sorted(billetera, reverse=True)
    sin_cambio = False    
    while amount > 0 and sin_cambio == False:        
        for valor in tipos_billete:            
            billetes_necesarios = int(amount/valor)            
            if billetera[valor] >= billetes_necesarios:
                cambio[valor] = billetes_necesarios
                amount = round(amount - (billetes_necesarios*valor),2)
            else:
                cambio[valor] = billetera[valor]
                amount = round(amount - (billetera[valor]*valor),2)            
            
        if amount > 0:
            sin_cambio = True
                
    if sin_cambio:        
        return None
    else:        
        return wallet(cambio)
```
