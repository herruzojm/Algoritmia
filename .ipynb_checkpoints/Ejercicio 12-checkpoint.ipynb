```python
# https://repl.it/Kdqt/1
import numpy as np

class OptimalBinarySearchTree:    
        
    def __init__(self, keys, probab_keys, probab_pseudokeys = None):        
        self.probab_keys = probab_keys
        if probab_pseudokeys == None:
            probab_pseudokeys=[0]*(len(probab_keys)+1)
        self.probab_pseudo = probab_pseudokeys
        self.keys = keys
        self.root_table = {}
        self.e = {}
        self.w = {}
        self.root_node = ""
        self.create_tables()        
        
    def create_tables(self):        
        n = len(self.probab_keys)
        
        for i in range(1, n + 2):
            self.e[(i, i - 1)] = self.probab_pseudo[i - 1]
            self.w[(i, i - 1)] = self.probab_pseudo[i - 1]
            
        for l in range(1, n + 1):
            for i in range(1, n - l + 2):
                j = i + l - 1
                self.e[(i, j)] = float("inf")
                self.w[(i, j)] = self.w[(i, j - 1)] + self.probab_keys[j - 1] + self.probab_pseudo[j]
                for r in range(i, j + 1):
                    t = round(self.e[(i, r - 1)] + self.e[(r + 1, j)] + self.w[(i, j)], 2)
                    if t < self.e[(i, j)]:
                        self.e[(i, j)] = t
                        self.root_table[(i, j)] = r
         
        self.root_node = self.keys[self.root_table[1,len(self.keys)]-1]
    
    def __len__(self):        
        return len(self.keys)        
        
    def __contains__(self, key):        
        return (key in self.keys)
        
    def __getitem__(self, i):        
        return self.keys[i]    

    def __str__(self, key = None):       
        if key == None:
            key = self.root()
            
        return self.string_tree(key)
          
    def string_tree(self, key = None):
        if key == None:
            return ""
        
        children = self.children(key)
        tree = "(" + self.string_tree(children[0]) + key + self.string_tree(children[1]) + ")"        
        return tree
            
    def children(self, key = None):        
        if key == None:            
            key = self.root_node
            
        root_index = self.keys.index(self.root_node)+1        
        left_index = (1, root_index-1)        
        right_index = (root_index + 1, len(self.probab_keys))
        right_node = self.keys[self.root_table[right_index]-1]
        left_node = self.keys[self.root_table[left_index]-1]
        
        if (key != self.root_node):
            current_key = self.root_node
            while current_key != key:
                               
                if self.keys.index(current_key) > self.keys.index(key):
                    i,j = left_index                   
                else:
                    i,j = right_index     
                    
                current_key = self.keys[self.root_table[i,j]-1]                
                raiz = self.keys.index(current_key)+1                
                
                left_index = (i, raiz - 1)
                right_index = (raiz + 1, j)                
                
                if left_index[1]>=left_index[0] and left_index in self.root_table:
                    left_node = self.keys[self.root_table[left_index]-1] 
                else:
                    left_node = None
                
                if right_index[1]>=right_index[0] and right_index in self.root_table:
                    right_node = self.keys[self.root_table[right_index]-1] 
                else:
                    right_node = None                                    
                                
        return (left_node, right_node)

    def depth(self, key = None):                            
        if key == None:
            return self.lenght(self.root())
        elif key not in self.keys:
            return None
        else:
            depth = 0
            current_key = self.root()

            while (current_key != key):
                depth = depth + 1
                children = self.children(current_key)
                    
                if self.keys.index(key) > self.keys.index(current_key):
                    current_key = children[1]
                else:
                    current_key = children[0]

            return depth
        
    def lenght(self, key):        
        if key == None:
            return -1
        else:
            children = self.children(key)            
            return 1 + max(self.lenght(children[0]), self.lenght(children[1]))

    def root(self):        
        return self.root_node        
        
    def cost(self, key=None):       
        if key == None or key == self.root():
            return self.e[1, len(self.keys)]
        else:
            #TODO
            print('TODO')
            return 1
        
    def print_matrix(self, data):
        matrix = np.zeros((len(self.keys)+1, len(self.keys)+1))
        for row, column in data:
            matrix[row-1][column-1] = data[row, column] 

        i = 0
        for row in matrix:
            row_lenght = len(row)-1        
            temp = np.insert(row, 0, row[row_lenght])        
            temp = np.delete(temp, row_lenght+1)        
            row = temp        
            matrix[i] = row
            i+=1
        print(matrix)        
        print()        

	
        
keys = ["k1", "k2", "k3", "k4", "k5"]
p = [0.15, 0.10, 0.05, 0.10, 0.20]
q = [0.05, 0.10, 0.05, 0.05, 0.05, 0.10]
tree = OptimalBinarySearchTree(keys, p, q)

print('e')
tree.print_matrix(tree.e)
print('w')
tree.print_matrix(tree.w)
print('root_table')
tree.print_matrix(tree.root_table)

```
